```python
# Q1. What is the mathematical formula for a linear SVM?
"""The mathematical formula for a linear SVM is defined by the linear kernel function: f(X) = w^T * X + b. 
In this equation, w is the weight vector that you want to minimize, X is the data that you’re trying to classify, 
and b is the linear coefficient estimated from the training data.  """
```

```python
# Q2. What is the objective function of a linear SVM?
"""The objective function of a linear SVM is to find a hyperplane in an N-dimensional space that distinctly classifies the data 
points . The goal is to find the maximum margin separating hyperplane . The margin is defined as the distance from the 
hyperplane to the closest point across both classes. """
```

```python
# Q3. What is the kernel trick in SVM?
"""The kernel trick in SVM is a technique used to transform nonlinearly separable data into a higher-dimensional feature 
space where the data becomes linearly separable. The kernel trick allows us to perform the SVM classification without 
actually computing the transformation explicitly

Linear kernel: The linear kernel is the simplest kernel function, and it is used when the data is already linearly 
separable or when there are a large number of features relative to the number of training examples. 
The linear kernel is defined as K(xᵢ, xⱼ) = xᵢ^T xⱼ, where xᵢ and xⱼ are the input vectors.

Polynomial kernel: The polynomial kernel is a popular kernel function that can capture nonlinearities in the data.
 It is defined as K(xᵢ, xⱼ) = (γ xᵢ^T xⱼ + r)^d, where γ, r, and d are hyperparameters. The polynomial kernel function
  computes the dot product of two vectors after mapping them to a higher-dimensional feature space using a polynomial function.

Gaussian (RBF) kernel: The Gaussian kernel is a widely used kernel function that can capture complex nonlinearities in the data. 
It is defined as K(xᵢ, xⱼ) = exp(-γ ||xᵢ - xⱼ||^2), where γ is a hyperparameter that controls the width of the Gaussian curve. 
The Gaussian kernel computes the dot product of two vectors after mapping them to an infinite-dimensional feature space using
 a Gaussian function
```

```python
# Q4. What is the role of support vectors in SVM Explain with example
"""Support vectors are the training samples that lie closest to the decision boundary or hyperplane in a support vector 
machine . In other words, support vectors are the data points that are used to define the margin of separation between
 the two classes in SVM.

The role of support vectors in SVM is crucial because they determine the location and orientation of the decision boundary.
 The support vectors are chosen because they are the most difficult points to classify, and hence they have the highest impact
  on the final classification.



"""
```

```python
# Q6. SVM Implementation through Iris dataset.

# Bonus task: Implement a linear SVM classifier from scratch using Python and compare its
# performance with the scikit-learn implementation.
# ~ Load the iris dataset from the scikit-learn library and split it into a training set and a testing setl
# ~ Train a linear SVM classifier on the training set and predict the labels for the testing setl
# ~ Compute the accuracy of the model on the testing setl
# ~ Plot the decision boundaries of the trained model using two of the featuresl
# ~ Try different values of the regularisation parameter C and see how it affects the performance of
# the model.
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC
import numpy as np
import matplotlib.pyplot as plt

iris = load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.2, random_state=42)
feature1 = 0
feature2 = 2
X_plot = X_train[:, [feature1, feature2]]
svm_clf = LinearSVC(random_state=42)
svm_clf.fit(X_plot, y_train)


x_min, x_max = X_plot[:, 0].min() - 0.5, X_plot[:, 0].max() + 0.5
y_min, y_max = X_plot[:, 1].min() - 0.5, X_plot[:, 1].max() + 0.5
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02), np.arange(y_min, y_max, 0.02))


Z = svm_clf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)


plt.contourf(xx, yy, Z, cmap=plt.cm.Spectral, alpha=0.8)
plt.scatter(X_plot[:, 0], X_plot[:, 1], c=y_train, cmap=plt.cm.Spectral)
plt.xlabel(iris.feature_names[feature1])
plt.ylabel(iris.feature_names[feature2])
plt.show()
```
